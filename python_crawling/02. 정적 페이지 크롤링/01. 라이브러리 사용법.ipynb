```python

# 원하는 주소를 가져오기
# 해당 페이지에 요청을 보내고(get), 페이지를 가져오기
#response : 응답 객체, 객체 = 데이터 + 명령어
response = requests.get("https://startcoding.pythonanywhere.com/basic")
response.status_code
#200 = 응답이 정상적으로 일어났음
#404 = 페이지를 찾을 수 없음

```

```python
html = response.text
# 문자열 형태로 출력되기 때문에, 원하는 데이터를 추출 하기 어렵다.
# >> beautifulsoup4 의 도움을 받아야 한다.


soup = BeautifulSoup(html, 'html.parser')
#문자열 형태의 html을, parser 객채로 하나하나 씩 잘라줌
soup

```

```python
soup.select_one(".brand-name")
#선택자를 만드는 과정이 정말 중요하다 !

# 크롤링 하려면, F12 를 누르면, 구글 개발자 도구를 열어준다 (= ctrl + shift + i , F12로 안열리시는 분 활용)

# 1. 테그의 별명이나, id 값 이 있는지 확인하기

soup.select_one(".brand-name").text
soup.select_one(".brand-name").attrs #주소 가져올때 사용
```
